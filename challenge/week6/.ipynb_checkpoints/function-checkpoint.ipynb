```python
import random

def get_number():
    return random.randrange(1, 46) # 1~45까지의 정수중에서 랜덤으로 생성

def set_lotto_number():
    lotto = [] # 빈 리스트 생성
    while len(lotto) <= 6:
        num = get_number() # 함수 호출 -> return값은 1~45중 하나이고 반환받은 값을 num에 저장
        if num not in lotto: # num이 lotto 리스트에 있지 않다면
            lotto.append(num) # num을 lotto 리스트에 추가
    return lotto
    
def match_numbers(user_inputs, lotto_numbers):
    same_numbers = (set(user_inputs) & set(lotto_numbers)) # user_inputs와 lotto_numbers중 겹치는 항목을 same_numbers에 자료형은 set로 저장
    return same_numbers

def check_prize(same_numbers):
    if len(same_numbers) == 6: # same_numbers의 길이(아이템 수)가 6이라면
        print('1st prize!')
    elif len(same_numbers) == 5: # same_numbers의 길이(아이템 수)가 5라면
        print('2nd prize!')
    elif len(same_numbers) == 4: # same_numbers의 길이(아이템 수)가 4라면
        print('3rd prize!')
    elif len(same_numbers) == 3: # same_numbers의 길이(아이템 수)가 3이라면
        print('4th prize!')
    else: # 1 또는 2인 경우
        print('try next week!')


num = 0

print('** Start Lotto **')
print('1) Set the lotto numbers')
lotto_numbers = set_lotto_number()

print('2) Set the user inputs')
user_inputs = [] # 빈 리스트 생성

while True:
    num = int(input('Enter %d-number: '%(len(user_inputs)+1))) # 초기 user_inputs의 길이(아이템 수)는 0이므로 +1
    if num < 1 or num > 46: # 입력값이 1보다 작거나 46보다 큰 경우 -> 1~46이 아닌 경우
        print('No valid number')
        continue # 반복문의 처음으로 이동
    if user_inputs.count(num) == 0: # user_inputs에 num의 갯수가 0이라면(=num이 없다면)
        user_inputs.append(num) # user_inputs에 num 추가
    else: # user_inputs에 num의 갯수가 0이 아니라면(=num이 있다면)
        print('Same number')
        continue # 반복문의 처음으로 이동
    if len(user_inputs) >= 6:
        break # 반복문 탈출

print('3) Match the numbers')
same_numbers = match_numbers(user_inputs, lotto_numbers)

print('4) Check the prize')
check_prize(same_numbers)

print('5) Sort the lists')
lotto_numbers.sort()
user_inputs.sort()

print('=== Summary ===')
print('Lotto numbers: ', *lotto_numbers) # *lotto_numbers: lotto_numbers의 아이템'만' 출력
print('User numbers: ', *user_inputs) # *user_inputs: user_inputs의 아이템'만' 출력
```
